```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('sample_data/california_housing_train.csv')
```

```python
data.describe()
```

```python
features = list(data)
```

```python
train_X = data[features[:len(features)-1]]
train_y = data['median_house_value']
```

```python
model = LinearRegression()
```

```python
model.fit(train_X,train_y)
```

```python
test_data = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
test_X = test_data[features[:len(features)-1]]
test_y = test_data['median_house_value']
```

```python
prediction = model.predict(test_X)
```

```python
print(mean_absolute_error(prediction,test_y))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model2 = DecisionTreeRegressor(random_state=1)
```

```python
model2.fit(train_X,train_y)
```

```python
prediction2 = model2.predict(test_X)
```

```python
print(mean_absolute_error(prediction2,test_y))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model3 = RandomForestRegressor(n_estimators = 50,random_state=1)
```

```python
model3.fit(train_X,train_y)
```

```python
prediction3 = model3.predict(test_X)
```

```python
print(mean_absolute_error(prediction3,test_y))
```

```python
from xgboost import XGBRegressor
```

```python
model4_np = XGBRegressor(random_state=1)
```

```python
model4_np.fit(train_X,train_y)
```

```python
prediction4_np = model.predict(test_X)
```

```python
print(mean_absolute_error(prediction4_np,test_y))
```

```python
model4 = XGBRegressor(n_estimators = 10000,learning_rate=0.005,n_jobs=4)
```

```python
model4.fit(train_X,train_y,early_stopping_rounds = 5,eval_set=[(test_X,test_y)],verbose = True)
```

```python
prediction4 = model4.predict(test_X)
```

```python
print(mean_absolute_error(prediction4,test_y))
```
